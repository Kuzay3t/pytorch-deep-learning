```python
# importing necessary libaries

import torch
import torch.nn as nn
import torch.optim as optim
```

```python
distances = torch.tensor([[1.0], [2.0], [3.0]], dtype = torch.float32)
times = torch.tensor([[6.96], [12.11], [16.77]], dtype = torch.float32)
```

```python
# creating the neural network

model = nn.Sequential(nn.Linear(1,1))

# creating the loss function and the optimizer
loss_function = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
# to create the training loop

for epoch in range(500):
  optimizer.zero_grad() # reset the optimizer
  outputs = model(distances) # make predictions
  loss = loss_function(outputs, times)
  loss.backward() # to calculate adjustment
  optimizer.step() # to adjust the weights and update the model
```

```python
with torch.no_grad():
  test_distance = torch.tensor([[25.0]], dtype=torch.float32)
  predicted_Time = model(test_distance)
  print(f"Predicted time for 25 miles: {predicted_Time.item():.1f} minutes")
```
